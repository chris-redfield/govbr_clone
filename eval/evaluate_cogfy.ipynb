```python
import pandas as pd
import os
from cogfy_manager import CogfyClient
from urllib.parse import urljoin
import requests
from tqdm.notebook import tqdm
from dotenv import load_dotenv
```

```python
load_dotenv()

df = pd.read_csv("../dataset/scripts_dataset.csv")
df
```

```python
k = os.environ["COGFY_API_KEY"]
api_url = "https://public-api.serpro.cogfy.com"
client = CogfyClient(base_url=api_url, api_key=k)
```

```python
# Collection Gov.br
collection_id = client.get_collection_id("Chat (https://wa.me/551151233964)")
collection_id
```

```python
client.find_collection(collection_id)
```

```python
field_id = "dcc473a3-6e98-4fd3-87c4-67d0091674f9"
chat_id = "5c400d65-b89d-4d93-801e-dbee18a122bb"
```

```python
client.find_chat(collection_id=collection_id, chat_id=chat_id)
```

```python
df["response_cogfy"] = ""

for idx, row in tqdm(df.iterrows(), total=len(df)):
    print(row.pergunta)
    print()
    response_create_chat = client.create_chat(
        collection_id=collection_id,
        field_id=field_id
    )
    chat_id = response_create_chat['id']

    response = client.send_chat_message(
        collection_id=collection_id,
        chat_id=chat_id,
        content=row.pergunta
    )
    # print(response)
    df.loc[idx, 'response_cogfy'] = response["content"]
```
